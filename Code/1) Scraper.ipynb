```python
import pandas as pd
import numpy as np
import re
import requests as req
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from fake_useragent import UserAgent
import os

%matplotlib inline
```

Defining the url we want to scrape.

```python
url = 'https://ingatlan.com/szukites/kiado+lakas+budapest'
```

Defining an agent that changes for each scraping. The data per page is 20 listings and we save them down as a soup object.

```python
ua = UserAgent()
header = {'User-Agent': ua.random}

listings_per_page = 20
page_addon = '?page='
    
r_ingatlan = req.get(url, headers=header)
soup = BeautifulSoup(r_ingatlan.content,"lxml")

listing_count = int(soup.find('div', {'class':'results__number'}).attrs['data-listings-count'].replace(' ',''))
```

```python
listing_count
```

The below is the basic setup for or scraping. Firs the agent is defined, the null "not defined" variables follow and for each page we start scraping for the listings. However, errors occur for some reason and that is where we chose to collect the failed listings so that we can scrape them later on.

```python
base_url = 'https://ingatlan.com/'

abc = []
errors = []


for pages in range(1, (listing_count // listings_per_page)+1):
    ua = UserAgent()
    header = {'User-Agent': ua.random}
        
    listing_url_page = urljoin(url,page_addon) + str(pages)
    r_ingatlan = req.get(listing_url_page, headers=header)
    soup = BeautifulSoup(r_ingatlan.content,"lxml")
     

    ingatlan_address = []
    
    for link in  soup.find_all('div', {'class':'listing__address'}):
        ingatlan_address.append(link.text)
    
    ingatlan_id = []
    
    for link in soup.find_all('button', {'class': 'listing__action--save button--link js-save-listing'}):
        ingatlan_id.append(link.get('data-id'))
        
        
    for i, k in zip(ingatlan_id, ingatlan_address):
        url_ingatlan = urljoin(base_url, i)
    
        ua = UserAgent()
        header = {'User-Agent': ua.random}
    
        r_ingatlan = req.get(url_ingatlan, headers=header)
        soup_ingatlan = BeautifulSoup(r_ingatlan.content,"lxml")
        
        ingatlan_quality = "not defined"
        ingatlan_yearbuilt = "not defined"
        ingatlan_floor = "not defined"
        ingatlan_floorsinbuilding = "not defined"
        ingatlan_lift = "not defined"
        ingatlan_height = "not defined"
        ingatlan_heating = "not defined"
        ingatlan_furnish = "not defined"
        ingatlan_balcony = "not defined"
        ingatlan_pet = "not defined"
        ingatlan_orientation = "not defined"
        ingatlan_view = "not defined"
        ingatlan_size = "not defined"
        ingatlan_price = "not defined"
        ingatlan_rooms = "not defined"
        ingatlan_newlybuilt = "not defined"
    
        try:
            for table in (soup_ingatlan.find('div', {'class':'card details'}).find('div', {'class':'paramterers'}).find_all('table')):

                for td in table.find_all('td'):
                    if td.text == 'Ingatlan állapota':
                        ingatlan_quality = td.findNextSibling().text
                    if td.text == 'Építés éve':
                        ingatlan_yearbuilt = td.findNextSibling().text
                    if td.text == 'Emelet':
                        ingatlan_floor = td.findNextSibling().text
                    if td.text == 'Épület szintjei':
                        ingatlan_floorsinbuilding = td.findNextSibling().text
                    if td.text == 'Lift':
                        ingatlan_lift = td.findNextSibling().text
                    if td.text == 'Belmagasság':
                        ingatlan_height = td.findNextSibling().text
                    if td.text == 'Fűtés':
                        ingatlan_heating = td.findNextSibling().text
                    if td.text == 'Bútorozott':
                        ingatlan_furnish = td.findNextSibling().text
                    if td.text == 'Erkély':
                        ingatlan_balcony = td.findNextSibling().text
                    if td.text == 'Kisállat':
                        ingatlan_pet = td.findNextSibling().text
                    if td.text == 'Tájolás':
                        ingatlan_orientation = td.findNextSibling().text
                    if td.text == 'Kilátás':
                        ingatlan_view = td.findNextSibling().text


                ingatlan_size = soup_ingatlan.find('div', {'class':'parameter parameter-area-size'}).find('span', {'class':'parameter-value'}).text
                ingatlan_price = soup_ingatlan.find('div', {'class':'parameter parameter-price'}).find('span', {'class':'parameter-value'}).text
                ingatlan_rooms = soup_ingatlan.find('div', {'class':'parameter parameter-room'}).find('span', {'class':'parameter-value'}).text
                ingatlan_newlybuilt = soup_ingatlan.find('span', {'class':'label label--newly-built'})
            
                abc.append([i, k, ingatlan_size, ingatlan_price, 
                           ingatlan_rooms, ingatlan_floor, ingatlan_quality, ingatlan_height, ingatlan_heating,
                            ingatlan_furnish, ingatlan_balcony, ingatlan_pet,
                           ingatlan_yearbuilt, ingatlan_floorsinbuilding, ingatlan_lift, ingatlan_orientation,
                           ingatlan_view])
                
        except:
            print("an error occured with", pages)
            errors.append(pages)
```

```python
pd.DataFrame(abc, columns = [
                'ID', 'Address', 'Size','Price','Rooms','Floor',
                'Quality', 'Inner Height', 'Heating', 'Furnished', 'Balcony', 'Pet', 
                'Year Built','Floors','Lift','Orientation','View'])
```
